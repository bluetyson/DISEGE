**First import Libraries. Note the disege library with the main functions**


```python
import matplotlib.pyplot as plt
#DISEGE library
from disege import processing,utils
```

**Set inclination and declination of the synthetic anomaly**

```python
inc, dec = -9.5, -20.4  # B = 24.76 - parameters at Amazonian terrain
```

**Define the dimensions of the rectangles following: (x_min, x_max, y_min, y_max, z_min, z_max). It's possible to insert several rectangles inside the brackets**

```python
syn_bounds = [(-750,750,-1500,1500,100,1100)]
```

**Insert the area of the computed anomaly, also following: (x_min, x_max, y_min, y_max, z_min, z_max)**

```python
area = (-2000,2000,-2000,2000,0,2000)
```

**Calculate The Total Magnetic Intesity (TMI) of the synthetic anomaly. Magnetization J in A/m**

```python
x,y,mag,model = utils.syn_data(inc, dec,syn_bounds,area,magnetization = 5)
```

**Grid the TMI. As default we use a 50-meter cell and the cubic interpolation method**

```python
xi,yi,z = utils.grid_data(x,y,mag,area,cel= 50, method = 'cubic')
```

**Plot the data. In this case the TMI**

```python
utils.syn_plot(xi,yi,z,area,syn_bounds,model,label='Total Magnetic Intensity',clabel='nT')
plt.show()
```

## The DISEGE processing class allow to apply several equations:

**Total Gradient or Analytic Signal (Nabighian 1972)**

```python
asa = processing.asa_mag(x,y,mag,area)
#Gridding and plotting again
xi,yi,z = utils.grid_data(x,y,asa,area)
utils.syn_plot(xi,yi,z,area,syn_bounds,model,label='Total Gradient or Analytic Signal (Nabighian 1972)',clabel='nT/m')
plt.show()
```

**Tilt Angle Filter - ISA (Miller and Singh, 1994)**

```python
tdr = processing.tdr_mag(x,y,mag,area)
#Gridding and plotting again
xi,yi,z = utils.grid_data(x,y,tdr,area)
utils.syn_plot(xi,yi,z,area,syn_bounds,model,label='Tilt Angle Filter (Miller and Singh 1994)',clabel='rad')
plt.show()
```

**Total Horizontal Derivative of the Tilt Angle - THD-TDR (Verduzco et al. 2004)**

```python
thd_tdr = processing.thd_tdr_mag(x,y,mag,area)
#Gridding and plotting again
xi,yi,z = utils.grid_data(x,y,thd_tdr,area)
utils.syn_plot(xi,yi,z,area,syn_bounds,model,label='Total Horizontal Derivative of the Tilt Angle (Verduzco et al. 2004)',clabel='rad/m')
plt.show()
```

**Theta Map (Wijns et al. 2005)**

```python
theta = processing.theta_mag(x,y,mag,area)
#Gridding and plotting again
xi,yi,z = utils.grid_data(x,y,theta,area)
utils.syn_plot(xi,yi,z,area,syn_bounds,model,label='Theta Map (Wijins et al. 2005)',clabel='')
plt.show()
```

**TDX Filter (Cooper and Cowan 2006)**

```python
tdx = processing.tdx_mag(x,y,mag,area)
#Gridding and plotting again
xi,yi,z = utils.grid_data(x,y,tdx,area)
utils.syn_plot(xi,yi,z,area,syn_bounds,model,label='TDX Filter (Cooper and Cowan 2006)',clabel='rad')
plt.show()
```

**Analytic Signal of Tilt Angle (Ansari and Alamdar 2011)**

```python
asta = processing.asta_mag(x,y,mag,area)
#Gridding and plotting again
xi,yi,z = utils.grid_data(x,y,asta,area)
utils.syn_plot(xi,yi,z,area,syn_bounds,model,label='Analytic Signal of Tilt Angle (Ansari and Alamdar 2011)',clabel='rad/m')
plt.show()
```

**Generalised Derivative Operator - GDO (Cooper and Cowan 2011)**

```python
gdo = processing.gdo_mag(x,y,mag,area,theta=0, phi=45)
#Gridding and plotting again
xi,yi,z = utils.grid_data(x,y,gdo,area)
utils.syn_plot(xi,yi,z,area,syn_bounds,model,label='Generalised Derivative Operator - GDO (Cooper and Cowan 2011)',clabel='')
plt.show()
```

**TAHG Filter (Ferreira et al. 2013)**

```python
tahg = processing.tahg_mag(x,y,mag,area)
#Gridding and plotting again
xi,yi,z = utils.grid_data(x,y,tahg,area)
utils.syn_plot(xi,yi,z,area,syn_bounds,model,label='TAHG (Ferreira et al. 2013)',clabel='rad')
plt.show()
```

**STDR Map (Nasuti et al. 2018)**

```python
stdr = processing.stdr_mag(x,y,mag,area)
#Gridding and plotting again
xi,yi,z = utils.grid_data(x,y,stdr,area)
utils.syn_plot(xi,yi,z,area,syn_bounds,model,label='STDR Map (Nasuti et al. 2018)',clabel='rad')
plt.show()
```

**Total Horizontal Derivative of STDR (Nasuti et al. 2018)**

```python
thd_stdr = processing.thd_stdr_mag(x,y,mag,area)
#Gridding and plotting again
xi,yi,z = utils.grid_data(x,y,thd_stdr,area)
utils.syn_plot(xi,yi,z,area,syn_bounds,model,label='Total Horizontal Derivative of STDR (Nasuti et al. 2018)',clabel='rad/m')
plt.show()
```
