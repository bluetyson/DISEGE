**First import Libraries. Note the disege library with the main functions**


```python
import matplotlib.pyplot as plt
#DISEGE library
from disege import processing,utils
```

**Load xyz data**

```python
x,y,z = utils.load_xyz(fname = 'MAG_cinz_reproj.xyz')
```

**Grid the data (cell size in meters)**

```python
xi,yi,zi = utils.grid_data(x,y,z,cel = 125)
```

**Plot the data**

```python
utils.plot_data(xi,yi,zi,label='Total Magnetic Intensity', clabel='nT', shaded=False)
plt.tight_layout()
plt.show()
```
